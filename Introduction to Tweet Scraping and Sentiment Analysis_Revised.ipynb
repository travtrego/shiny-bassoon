```python
from twarc import Twarc2, expansions
import datetime
import json
import csv

# Replace your bearer token below. You may use below code if you do not have a twitter developer account yet
client = Twarc2(bearer_token="AAAAAAAAAAAAAAAAAAAAAGSXZQEAAAAAk6YVv7DVW74eDZZ5u698Inv4QXc%3DCpTroI60R3imjzDPgwoEr6CRWIWjxoqSR1jOTQJtl0to45EHce")


def main():
    # Specify the start time in UTC for the time period you want replies from
    start_time = datetime.datetime(2022, 3, 20, 0, 0, 0, 0, datetime.timezone.utc)

    # Specify the end time in UTC for the time period you want Tweets from
    end_time = datetime.datetime(2022, 3, 21, 0, 0, 0, 0, datetime.timezone.utc)

    # This is where we specify our query as discussed in module 5.
    query = "(#climatechange, #globalwarming) OR (climate change) -is:retweet lang:en place:Boston"

    # Name and path of the file where you want the Tweets written to
    file_name = 'data\climateChange_20220320.csv'

    # The search_all method call the full-archive search endpoint to get Tweets based on the query, start and end times
    search_results = client.search_all(query=query, start_time=start_time, end_time=end_time, max_results=100)

    # Twarc returns all Tweets for the criteria set above, so we page through the results
    for page in search_results:
        # The Twitter API v2 returns the Tweet information and the user, media etc.  separately
        # so we use expansions.flatten to get all the information in a single JSON
        result = expansions.flatten(page)
        # We will open the file and append one JSON object per new line
        with open(file_name, 'a', newline='', encoding="utf8") as filehandle:
            writer = csv.writer(filehandle, delimiter='|')
            for tweet in result:
                d={}
                oneTweet=json.dumps(tweet)
                onetweet=json.loads(oneTweet)
                d['tweet_id'] = onetweet['id']
                d['created_at'] = onetweet['created_at']
                d['user_name']=onetweet['author']['username']
                d['text']=onetweet['text'].replace('\n', ' ').replace('\r', '')
                #print(d)
                writer.writerow(list(d.values()))

if __name__ == "__main__":
    main()
```

```python
import pandas as pd

climate=pd.read_csv('data\climateChange_20220320.csv',sep="|", header=None)

cols=['tweet_id', 'created_at', 'user_name', 'text'] # define columns header

climate.columns=cols
```

## [Sentiment Analysis](https://pypi.org/project/vaderSentiment/)

```python
#! pip install vaderSentiment==3.3.2
```

```python
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer

sentiment = []
analyzer = SentimentIntensityAnalyzer()

for index, row in climate.iterrows():
    tweet = str(row['text'])
    vs = analyzer.polarity_scores(tweet)
    sentiment.append(vs)
    
climate['VaderSentiment'] = sentiment
```

```python
def sent_classify(score):
    if score>0:
        return "Positive"
    elif score<0:
        return "Negative"
    else:
        return "Neutral"
```

```python
climate['sentiment']=climate['compound'].apply(sent_classify)
```

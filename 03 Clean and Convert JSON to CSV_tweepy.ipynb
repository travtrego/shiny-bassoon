```python
import json

# read tweets by line and store the result into a list
data = []

fin = open('climate_change.json', 'r', encoding="utf8")
Lines = fin.readlines()

# Occasionally the JSON for a tweet is messed up, this confirms that the tweet is in valid JSON format
for line in Lines:
    try:
        data.append(json.loads(line))
    except:
        print("Skipping")
```

```python
# selected certain fields for further analysis and store it into a csv file. The saved csv file has no header

import csv

with open('climate_change.csv', 'w', newline='', encoding="utf8") as file:
    writer = csv.writer(file, delimiter='|')
    for tweet in data:
        d = {}
        d['tweet_id'] = tweet['id']
        d['created_at'] = tweet['created_at']
        d['source']=tweet['source']
        d['lang']=tweet['lang']
        d['user_name']=tweet['user']['screen_name'].replace('\n', ' ').replace('\r', '')
        d['name']=tweet['user']['name']
        # not all tweets have location, 
        if tweet['user']['location'] is not None:
            d['user_location']=tweet['user']['location'].replace('\n', ' ').replace('\r', '')
        else:
            d['user_location']=None
        d['user_followers_count']=tweet['user']['followers_count']
        d['user_friends_count']=tweet['user']['friends_count']
        if'retweeted_status' in tweet.keys():
            d['retweet_created_at']=tweet['retweeted_status']['created_at']
            d['retweet_user_name']=tweet['retweeted_status']['user']['screen_name'].replace('\n', ' ').replace('\r', '')
            d['retweet_name']=tweet['retweeted_status']['user']['name'].replace('\n', ' ').replace('\r', '')
        else:
            d['retweet_created_at']=None
            d['retweet_user_name']=None
            d['retweet_name']=None
        d['text']=tweet['text'].replace('\n', ' ').replace('\r', '')
        writer.writerow(list(d.values()))
```
